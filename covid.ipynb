```python
import pandas as pd
```

```python
pd.set_option("display.max_rows", 10)
confirmed_raw = pd.read_csv("confirmed.csv")
```

```python
confirmed_raw = pd.read_csv("confirmed.csv")
death_raw = pd.read_csv("death.csv")
recovered_raw = pd.read_csv("recovered.csv")
```

```python
# remove lat/long columns
confirmed_raw.drop(["Lat", "Long"], axis=1, inplace=True)
death_raw.drop(["Lat", "Long"], axis=1, inplace=True)
recovered_raw.drop(["Lat", "Long"], axis=1, inplace=True)
```

```python
# Keep only countries (sum the regions)
confirmed_by_country = confirmed_raw.groupby("Country/Region").sum().transpose()
death_by_country = death_raw.groupby("Country/Region").sum().transpose()
recovered_by_country = recovered_raw.groupby("Country/Region").sum().transpose()
```

```python
#convert datetime
confirmed_by_country.index = pd.to_datetime(confirmed_by_country.index, format='%m/%d/%y')
death_by_country.index = pd.to_datetime(death_by_country.index, format='%m/%d/%y')
recovered_by_country.index = pd.to_datetime(recovered_by_country.index, format='%m/%d/%y')
```

```python
# create active cases dataset
active_by_country = confirmed_by_country.sub(death_by_country).sub(recovered_by_country)
```

```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource, HoverTool
from ipywidgets import interact
```

```python
def show_graph(title, y_label, data):
    # output to static HTML file
    output_notebook()

    # hover tooltip
    hover = HoverTool(
        tooltips=[("name", "$name"), ("date", "$x{%F}"), ("value", "$y")],
        formatters={"$x": "datetime"},
    )
    # create a new plot
    p = figure(
        x_axis_type="datetime",
        tools="pan,box_zoom,reset,save",
        title=title,
        x_axis_label="date",
        y_axis_label=y_label,
        plot_width=1200,
        plot_height=700,
    )
    p.add_tools(hover)

    source = ColumnDataSource(data)

    # add some renderers
    l = p.line(
        name="France",
        x="index",
        y="France",
        source=source,
        legend_label="France",
        line_width=3,
        line_color="Blue",
    )
    l = p.line(
        name="Italy",
        x="index",
        y="Italy",
        source=source,
        legend_label="Italy",
        line_width=3,
        line_color="DarkGreen",
    )
    l = p.line(
        name="US",
        x="index",
        y="US",
        source=source,
        legend_label="US",
        line_width=3,
        line_color="Black",
    )
    l = p.line(
        name="Canada",
        x="index",
        y="Canada",
        source=source,
        legend_label="Canada",
        line_width=3,
        line_color="Red",
    )

    p.legend.location = "top_left"
    # show the results
    show(p)


def show_confirmed(ma=1):
    show_graph(
        "Confirmed cases by country",
        "confirmed cases",
        confirmed_by_country.rolling(ma).mean(),
    )


def show_confirmed_speed(ma=1):
    show_graph(
        "Confirmed cases speed by country",
        "new confirmed cases per day",
        confirmed_by_country.diff().rolling(ma).mean(),
    )


def show_death(ma=1):
    show_graph("Death by country", "deaths", death_by_country.rolling(ma).mean())


def show_death_speed(ma=1):
    show_graph(
        "Death speed by country",
        "new deaths per day",
        death_by_country.diff().rolling(ma).mean(),
    )


def show_recovered(ma=1):
    show_graph(
        "Recovered by country",
        "recovered cases",
        recovered_by_country.rolling(ma).mean(),
    )


def show_recovered_speed(ma=1):
    show_graph(
        "Recovered speed by country",
        "new recovered cases per day",
        recovered_by_country.diff().rolling(ma).mean(),
    )


def show_active(ma=1):
    show_graph(
        "Active cases by country", "active cases", active_by_country.rolling(ma).mean()
    )


def show_active_speed(ma=1):
    show_graph(
        "Active cases speed by country",
        "diff active cases per day",
        active_by_country.diff().rolling(ma).mean(),
    )
```
